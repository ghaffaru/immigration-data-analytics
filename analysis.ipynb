```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
# changing the index values of Haiti to type integer for plotting
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# Annotating
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

# annotate the 2010 earthquake
plt.text(2010, 6000, '2010 Earthquake')
```

```python
# compare the number of immigrants from India and China from 1980 to 2013
chinaIndia = df_can.loc[['China', 'India'], years]
```
